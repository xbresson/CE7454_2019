### Reshaping a tensor

```python
x=torch.arange(10)
print(x)
```

```python
print( x.view(2,5) )
```

```python
print( x.view(5,2) )
```

### Note that the original tensor x was NOT modified

```python
print(x)
```

### To make the change permanent you need to create a new tensor

```python
y=x.view(5,2)

print(x)
print('')
print(y)
```
